```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#from sklearn.
```

```python
dataset = pd.read_csv("/content/drive/MyDrive/Final Project 1.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
x_data=dataset.iloc[:,:-1]
y_data=dataset.iloc[:,-1]
```

```python
st_Scale = StandardScaler()
x_data = st_Scale.fit_transform(x_data)
```

```python
x_data=pd.DataFrame(x_data,columns=dataset.columns[: -1])
```

```python
x_data=dataset.iloc[:,:-1]
y_data=dataset.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.3,shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import *
dt_model = DecisionTreeClassifier()
model = dt_model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *
rfc_model = RandomForestClassifier()
model = rfc_model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import *
abc_model = AdaBoostClassifier()
model = abc_model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```
